```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
```

```python
img=cv2.imread('./images.jpeg')

plt.imshow(img) #BGR
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
#firstly it would download pretrained model
predictions=DeepFace.analyze(img) 
predictions
```

```python
predictions['dominant_emotion']
```

```python
faceCascade=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python

faces=faceCascade.detectMultiScale(gray,1.1,6)
```

```python
# draw a rectangle around the faces
for(x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
font=cv2.FONT_HERSHEY_SIMPLEX
# use puttext() methode for inserting text on video
cv2.putText(img, predictions['dominant_emotion'],(0,50),font,3 ,(0,0,255),3,cv2.LINE_4);
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
